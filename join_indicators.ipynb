```python
import pandas as pd
```

```python
# import data
ind = {}
for i in range(22):
    ind["ind_t%s" % str(i+1)] = pd.read_csv("ind_t%s.csv" % str(i+1))
```

```python
# join data
join_ind = ind['ind_t1']
for i in range(2,23):
    join_ind = join_ind.join(ind['ind_t%s' %str(i)].iloc[:,5:])
```

```python
join_ind.shape
```

```python
# check missing values for each column
join_ind.isnull().sum().sort_values(ascending=False)
```

```python
# output dataframe to csv
join_ind.to_csv('joined_indicators.csv')
```

```python
# group by district and sum up the section feature values for each district
df = join_ind.groupby(['indicadores_t1.cod_ccaa', 'indicadores_t1.cod_prov','indicadores_t1.cod_muni','indicadores_t1.distrito']).sum()
```

```python
# delete the column of section number
df = df.drop('indicadores_t1.seccion', axis=1)
```

```python
# output dataframe to csv
df.to_csv('joined_indicators_groupby_district.csv')
```
